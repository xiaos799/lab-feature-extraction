```python
num = customer_df[['customer_lifetime_value', 'income', 'monthly_premium_auto', 'months_since_last_claim', 'months_since_policy_inception', 'number_of_open_complaints', 'number_of_policies', 'total_claim_amount']]
cat = customer_df[['state', 'response', 'coverage', 'education', 'effective_to_date', 'employmentstatus', 'gender', 'location_code', 'marital_status', 'policy_type', 'policy', 'renew_offer_type', 'sales_channel', 'vehicle_class', 'vehicle_size','months_since_last_claim', 'months_since_policy_inception',
       'number_of_open_complaints', 'number_of_policies']]
```

```python
for x in ['customer_lifetime_value', 'income', 'monthly_premium_auto', 'total_claim_amount']:
    plt.figure(figsize=(10,5))
    plt.xlim(0,2000)
    plt.xlabel(x)
    sns.kdeplot(customer_df[x],shade=True)
    plt.show()
```
